### Cleaning up the data

```python
import pandas as pd
#loading the files drom the data folder
with open(r"C:\Users\ericg\Downloads\DATA\DATA\TABLES\Uganda_Karamoja_District_Crop_Yield_Population.csv") as a:
    df_a = pd.read_csv(a)
df_a
# print(df.head())
```

```python
#loading the files
with open(r"C:\Users\ericg\Downloads\DATA\DATA\TABLES\Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv")as b:
    df_b = pd.read_csv(b)
df_b
```

```python
df_b.shape
```

```python
#it seems that the data set is split into different folders and tables hence i have to combine the data and do 'concantination'
#data frame a and b have different collumn numbers
```

```python
data_frames = [df_a, df_b]  #creating them into a list to make concatation easier
concated_data_frame = pd.concat(data_frames)
concated_data_frame #the joined dataframe
```

```python
#cleaning up the data by reducing the number of decimal places tot two 
 #converting the dataframe to intagers as the round function does not apply to strings
numeric_cols = concated_data_frame.select_dtypes(include=['number']).columns
#rounding up the data in the dataframe
concated_data_frame[numeric_cols] = concated_data_frame[numeric_cols].apply(lambda x: round(x, 2))
#returning the first five rows in the dataframe
concated_data_frame.head()
```

```python
#'NAME', collumns mostly seem to be empty
#SUBCOUNTY_NAME, DISTRICT_NAME, Karamoja  apear to have missing values 
```

```python
# confirming column names
concated_data_frame.columns
```

```python
#confirming the shape of the dataframe
concated_data_frame.shape
```

```python
#lets arrange the dataframe in ascending order with refrence to 'object id'
sorted_concat_df = concated_data_frame.sort_values(by='OBJECTID', ascending=True)
sorted_concat_df

#we are sorting the data to deteramin if there are missing valuse according to 'OBJECTID' 
```

```python
#how many subcounties are in the dataset
print(sorted_concat_df['SUBCOUNTY_NAME'].unique())
#how many subcounties are in the dataset
print(f"There are {len(sorted_concat_df['SUBCOUNTY_NAME'].unique())} subcounties")
```

```python
#how many district names are in the dataset
print(sorted_concat_df['DISTRICT_NAME'].unique())
#how many districts are in the dataset
print(f"There are {len(sorted_concat_df['DISTRICT_NAME'].unique())} districts")
```

```python
#the OBJECTID column seems to be missing data
```

```python
# Drop the column 'NAME' from the DataFrame as most of the cells are empty
sorted_concat_df = sorted_concat_df.drop(columns=['NAME'])
sorted_concat_df

```

```python
#to remove the rows with empty values as the empty values such as SUBCOUNTY_NAME, DISTRICT_NAME and Karamoja we can not assume redion and are a vital part of data
sorted_concat_df.dropna(inplace = True) 
sorted_concat_df
```

### is there a cooorelation btw population and crop area hactre?

```python
import numpy as np
import pandas as pd
from scipy.stats import spearmanr

# Calculate Spearman's rank correlation coefficient and p-value
correlation, p_value = spearmanr(sorted_concat_df['Area'], sorted_concat_df['Crop_Area_Ha'])

# Print the results
print(f"Spearman's rank correlation coefficient: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")
```
